```python
import pandas as pd
import matplotlib.pyplot as plt
import star_trek_tos_data as TOS
```

```python
df = pd.DataFrame(TOS.main_cast_movies).T.set_index('original_title')[['count', 'main_cast_ids', 'main_cast_names', 'vote_average', 'vote_count']]
df = df[df['vote_count']>25]
df = df[df['count']>1]
df.sort_values('vote_average', ascending = False)
```

```python
df['vote_average'] = df['vote_average'].astype(float)
df.groupby('count').mean().plot( kind = 'bar')
plt.show()
```
