```python
#pull in the saved data for main cast total credits

import json

with open('main_credits.txt', 'r') as file:
    main_credits = json.load(file)
    main_credits = {int(id_str): main_credits[id_str] for id_str in main_credits}
    
main_ids=["1748","1749","1750","1751","1752","1753","1754","1755","1214943","1214944"]

with open('main_cast_stars.txt', 'r') as file:
    main_stars = json.load(file)
    main_stars = {int(id_str): main_stars[id_str] for id_str in main_stars}
```

```python
import pandas as pd
```

```python
#set up the dataframe to hold the main cast data
data_df= {}
data_df=pd.DataFrame(data_df)
data_df["Name"]= ' '
data_df["Job"]= ' '
data_df["Media"]= ' '
data_df["Date"]= ' '
data_df["Genre"]= ' '
data_df["Title"]= ' '
data_df
```

```python
#load selected data into the dataframe
counter=0
for id in main_ids:
    act_credits=len(main_credits[int(id)]['cast'])
    crew_credits=len(main_credits[int(id)]['crew'])    
    
    for x in range(0, act_credits):
        data_df.set_value(counter,"Name",main_stars[int(id)]['name'] )
        data_df.set_value(counter,'Job','Actor')
        data_df.set_value(counter,'Media',main_credits[int(id)]['cast'][x]['media_type'])
        data_df.set_value(counter,'Genre',main_credits[int(id)]['cast'][x]['genre_ids'])
        if main_credits[int(id)]['cast'][x]['media_type']=="movie":
            data_df.set_value(counter,'Date',main_credits[int(id)]['cast'][x]['release_date'])
            data_df.set_value(counter,'Title',main_credits[int(id)]['cast'][x]['title'])
            
        else:
            data_df.set_value(counter,'Date',main_credits[int(id)]['cast'][x]['first_air_date'])
            data_df.set_value(counter,'Title',main_credits[int(id)]['cast'][x]['name'])
        counter= counter + 1
            
    for x in range(0,crew_credits):
        data_df.set_value(counter,"Name",main_stars[int(id)]['name'])
        data_df.set_value(counter,'Job',main_credits[int(id)]['crew'][x]['job'])
        data_df.set_value(counter,'Media',main_credits[int(id)]['crew'][x]['media_type'])
        data_df.set_value(counter,'Genre',main_credits[int(id)]['cast'][x]['genre_ids'])
        if main_credits[int(id)]['crew'][x]['media_type']=="movie":
            data_df.set_value(counter,'Date',main_credits[int(id)]['crew'][x]['release_date'])
            data_df.set_value(counter,'Title',main_credits[int(id)]['crew'][x]['title'])
            
        else:
            data_df.set_value(counter,'Date',main_credits[int(id)]['crew'][x]['first_air_date'])
            data_df.set_value(counter,'Title',main_credits[int(id)]['crew'][x]['name'])
        counter= counter + 1
data_df = data_df.reset_index(drop=True)            
data_df
```

```python
#read the genre csv file and load the DataFrame
genres="genre_list.csv"
genre_dics=pd.read_csv(genres,encoding="iso-8859-1")

#genre_dics
```

```python

activity=data_df.groupby(['Name', 'Job', 'Media']).size().reset_index(name="count")
activity.head()

```

```python

test=activity.pivot_table( index= "Name", columns = "Job", values = "count")
test.fillna(0)
test
```

```python
test["Characters"]=test['Characters'].fillna(0)
test["Creator"]=test['Creator'].fillna(0)
test["Director"]=test['Director'].fillna(0)
test["Executive Producer"]=test['Executive Producer'].fillna(0)
test["Novel"]=test['Novel'].fillna(0)
test["Producer"]=test['Producer'].fillna(0)
test["Writer"]=test['Writer'].fillna(0)
test
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use('ggplot')
```

```python
# create the bar chart of activity
acting_gigs = {}

fig, ax = plt.subplots()


x_axis = np.arange(0, len(test) )
tick_locations = []
for x in x_axis:
    tick_locations.append(x + 0.4)
plt.xlim(-0.25, len(test))
plt.ylim(min(test['Actor']) - 1, max(test['Actor']) + 5)
plt.title(" Acting jobs for original cast of Star Trek")
plt.ylabel("Number of Jobs")

plt.xticks(tick_locations, test.index, rotation= "vertical")
bar_chart=plt.bar(x_axis, test['Actor'], color = 'r', alpha=0.75, align="edge")

   

plt.savefig("Acting_Jobs_Main_Cast.png")
plt.show()
```
