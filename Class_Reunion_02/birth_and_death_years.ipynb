```python
guest_death_years = []
for star in guest_stars:
    print(star, '-' ,guest_stars[star]['name'] , end = '\t')
    try:
        year = guest_stars[star]['deathday']
        print(year)
        if len(str(year)) > 5:
            guest_death_years.append(int(year[0:4]))
    except KeyError:
        print("hasn't died")
```
